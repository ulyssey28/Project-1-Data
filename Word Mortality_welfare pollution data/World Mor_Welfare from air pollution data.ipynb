```python
#read file 
import pandas as pd
filepath = "EXP_MORSC_01032019225444048.csv"
country_mort_data = pd.read_csv(filepath)
country_mort_data.head()
```

```python
# country_mort_data['Country'].value_counts()
```

```python
#Filter data for data relevant to the US.
US_data = country_mort_data[country_mort_data['Country'].str.contains("United States")==True]
US_data
```

```python
US_data.head()
```

```python
US_data_mix_var = US_data[['Country', 'Variable', 'Year', 'Unit', 'Value']]
US_data_mix_var.head(5)
# US_data_analysis['Year'].value_counts()
```

```python
#filter the "United states" data frame for variable values pertaining to mortality
Mort_data_analysis = US_data_mix_var.loc[US_data_mix_var['Variable'] == "Mortality from exposure to outdoor PM2.5", :]
#Note at this point the data still contains values for United States virgin Islands
#So we filter out the Virgin Islands and get Data for only United States 
US_PM25_Mort_data = Mort_data_analysis.loc[Mort_data_analysis['Country'] == "United States", :]


US_PM25_Mort_data.sort_values(by=['Year']).head()
```

```python
US_PM25_Mort_data.to_csv("../DataFrames/US_PM25_Mort_data.csv", index=False, header=True)
```
